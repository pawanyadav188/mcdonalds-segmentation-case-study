## üß† Step 3: Fermi Estimation ‚Äì Estimating Early Market Size for E-Scooters in Karnataka

Fermi Estimation is a method to logically break down a market size using basic assumptions, especially when full data isn‚Äôt available. Based on the FY 2024‚Äì25 EVreporter report and government datasets, this step estimates how many people in Karnataka (especially Bangalore) are likely to adopt e-scooters in the early market.

---

### üìà Why Focus on Electric Scooters (E-2W)?

As per the **EVreporter FY 2024‚Äì25 report**:
- **12,09,772 electric two-wheelers (E-2W)** were sold in India ‚Äî making up **59.4%** of total EV sales.
- E-scooters far outperformed EV cars (1,15,800 units).
- Two-wheelers dominate urban India due to low cost, personal convenience, and daily commuting needs.
- Government subsidies and rising fuel prices are further boosting demand.

This confirms that **e-scooters are the most commercially viable EV product for mass adoption**.

---

### üìç Why Karnataka?

From combined data:
- Karnataka is the **2nd highest state** in E-2W sales (1,48,304 units sold).
- EV penetration in Karnataka‚Äôs 2W segment is **11.3%**, compared to **national avg of 6.2%**.
- Karnataka has **India‚Äôs largest EV charging network** with **5,765 public charging stations**.
- Bangalore, the capital, is a tech hub with high population density, daily commuters, and early adopters.

All signs point to Karnataka ‚Äî particularly **Bangalore** ‚Äî as the most suitable early market for EV launch.

---

### üßÆ Fermi Estimation ‚Äì Market Size in Bangalore

| Parameter                                    | Estimate               | Justification |
|---------------------------------------------|------------------------|----------------|
| Total population of Bangalore               | 20 million             | 2024 estimate |
| % aged 18‚Äì45 (working segment)              | 45% ‚Üí 9 million        | Commuting age group |
| % who commute daily                         | 60% ‚Üí 5.4 million      | Urban mobility users |
| % with commute < 30 km/day                  | 65% ‚Üí 3.5 million      | Ideal range for e-scooters |
| % interested in EVs (early adopters)        | 10% ‚Üí 3,50,000         | Based on urban awareness |
| % who can afford e-scooter this year        | 5% ‚Üí **17,500 buyers** | ~‚Çπ90,000‚Äì‚Çπ1,00,000 price range |

---

### ‚úÖ Final Estimate:

> Based on Fermi logic, **17,000 to 18,000 people in Bangalore** are realistic e-scooter buyers in the next 12 months.

This number can easily grow to **40,000+ buyers** across Karnataka by including Tier-2 cities like Mysuru, Mangaluru, and Hubballi.

---

### üéØ Business Relevance:

This estimation supports the startup‚Äôs decision to:
- Choose **electric scooters** over other EV types

  - High adoption rate

## üîç Step 6: Segmentation Using Clustering (K-Means)

To uncover patterns across Indian states, K-Means clustering was applied using two features:
1. Percentage of EVs in total vehicle sales (`EV_Share`)
2. Number of public charging stations (`Charging_Stations`)

Clustering helps group states into meaningful categories like:
- Early adopters with strong infrastructure
- Low-adoption but high potential states
- Lagging regions with poor support

This technique helps the startup **prioritize regions** for investment, marketing, and partnerships.


````python
# Merge both datasets on 'State'
merged_df = pd.merge(
    ev_state_df,
    ev_charging_df.rename(columns={'State/ UT': 'State'}),
    on='State',
    how='inner'
)

# Clean and rename for clustering
merged_df.rename(columns={
    '% of Share of EV in Total Vehicles Sold': 'EV_Share',
    'No. of PCS as on 31st March 2024': 'Charging_Stations'
}, inplace=True)

# Keep only numeric columns needed
# Corrected
cluster_data = merged_df[['EV_Share', 'PCS_Count']].copy()

````

````python
from sklearn.cluster import KMeans

# Create model and fit with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
merged_df['Cluster'] = kmeans.fit_predict(cluster_data)
````

````python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=merged_df,
    x='PCS_Count',         # Use actual existing column name
    y='EV_Share',
    hue='Cluster',
    palette='Set1',
    s=100
)
plt.title("Clustering of States by EV Share and Charging Infrastructure")
plt.xlabel("Number of Public Charging Stations")
plt.ylabel("EV Share (%)")
plt.legend(title='Cluster')
plt.tight_layout()
plt.grid(True)
plt.show()
````
